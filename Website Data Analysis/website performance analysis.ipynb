```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions" ,"Users"]].sum().plot(ax=plt.gca())
plt.title('Sessions and users over time')
plt.xlabel('DateHour')
plt.ylabel('count')
plt.show()
plt.show()
```

# total users by channel

```python

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='channel group', y='Users', estimator=np.sum, palette='viridis')
plt.title('Total user by channel')
plt.xticks(rotation=45)
plt.show()
```

# Average engagement time by channel

```python

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='channel group', y='Average engagement time per session', estimator=np.mean, palette='magma')
plt.title("Avg Engagement Time by channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs Non Engaged session

```python
df.head()
```

```python
# Group data and calculate total sessions
# Group data and calculate total sessions
session_df = df.groupby("channel group")[["Sessions", "Engaged Sessions"]].sum().reset_index().rename(columns={"channel group": "Channel Group"})

# Calculate non-engaged sessions
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged Sessions"]

# Unpivot (melt) the DataFrame for visualization
session_df_melted = session_df.melt(id_vars="Channel Group", value_vars=["Engaged Sessions", "Non-Engaged"])

# Plot the clustered bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x="Channel Group", y="value", hue="variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()

```

# traffic by hour and channel

```python
heatmap_data =df.groupby(["Hour" ,"channel group" ,])["Sessions"].sum() .unstack().fillna(0)

plt.figure(figsize=(12 ,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5, annot=True, fmt=".0f")
plt.title("Traffic by Hour and channel")
plt.xlabel('Channel Group')
plt.ylabel("Hour of Day")
plt.show()
```

```python
# Group data by DateHour and calculate the average
# Group data by DateHour and calculate the average
df_plot = df.groupby("DateHour")[["Engagement rate ", "Sessions"]].mean().reset_index().rename(columns={"Engagement rate ": "Engagement Rate"})

# Plot both metrics on the same chart
plt.figure(figsize=(10, 5))
plt.plot(df_plot["DateHour"], df_plot["Engagement Rate"], label="Engagement Rate", color="green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()

```
